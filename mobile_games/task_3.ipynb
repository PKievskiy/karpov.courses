# Задача:
В игре Plants & Gardens каждый месяц проводятся тематические события, ограниченные по времени. В них игроки могут получить уникальные предметы для сада и персонажей, дополнительные монеты или бонусы. Для получения награды требуется пройти ряд уровней за определенное время. С помощью каких метрик можно оценить результаты последнего прошедшего события?

Предположим, в другом событии мы усложнили механику событий так, что при каждой неудачной попытке выполнения уровня игрок будет откатываться на несколько уровней назад. Изменится ли набор метрик оценки результата? Если да, то как?

## Часть 1: какие метрики помогут оценить результаты последнего ивента

В целом подобные события, по моему мнению, проводятся для привлечения новых пользователей и чтобы спровоцировать уже играющих пользователей на трату бонусов/бустов (и соответственно покупку) для получения уникального контента. 

А значит, укрупнённо самым важным для нас будет рост активности аудитории (и размера самой аудитории) приложения и показатели дохода в период события.

Стоит проанализовать как специфичные метрики самого ивента, так и метрики продукта в целом. 

При выборе метрик я исходил из того, что уровни в ивенте - отдельная специальная часть игры, которая не входит в основной игровой трек. На её прохождение отводится несколько дней.

## Метрики ивента

Все метрики последнего ивента стоит сравнивать с показателями предыдущих ивентов + с метриками игры, когда ивенты не проводятся. 

### Метрики активности аудитории

   

   • Количество игроков, принявших участие в последнем ивенте.  
   • DAU/WAU в ивенте (а не в игре в целом)  
   • Retention в ивенте (при анализе важно учитывать, что когорта пользователей, которые быстро прошли все уровни ивента, не будет возвращаться в него повторно).  

Все эти метрики помогут понять, какой доле от аудитории приложения ивент был интересен и насколько (удерживал ли ивент пользователей в течение всего времени проведения).  


### Метрики игровые

   • Распределение по: 
1. пройденным уровням на конец ивента (условно первый уровень прошли 5000 тыс. игроков, второй - 4000,  третий - 2000,  четвертый -1000 и т. д.)
2. количеству попыток на прохождение каждого уровня.
3. времени, затрачиваемому на каждый уровень.
4. количеству (сумме) покупок на каждом уровне.
5. количеству использованных бонусов на прохождение уровня.

Метрики берём в различных разрезах аудитории (в среднем по всем, в зависимости от уровня в основной игре, от количества потраченных бонусов на прохождение т.д.). Эти метрики помогут нам понять, насколько ивент был сложен/легок для пользователей, где возникли основные трудности и как они повлияли на игроков (стимулировали их покупать бусты или наоборот забросить ивент и т.д.)

   Воронка событий для детального анализа первой сессии на новых уровнях. Поможет понять всё ли ОК. Мб часть пользователей отваливаются по каким-то техническим причинам, т.к. эта часть продукта новая.
   Статистика использования новых предметов, персонажей. Пользуются ли они популярностью у игроков?

## Метрики продукта

### Метрики активности аудитории

   • DAU/WAU в продукте в целом в период проведения ивента. Увеличивает ли ивент активность аудитории? Привлекает ли он игроков, которые неактивно играют в основную часть игры, или в событии принимают участие +/- те же люди, что и обычно заходят в игру.  
   • Время, проведённое пользователями в игре. Стали ли люди во время ивента играть больше?  
   • Средняя продолжительность сессии (увеличивается ли она по сравнению с длиной сессии без ивентов?)  
   • Конверсия в участие в ивенте после просмотра пуша/рекламы + метрики, показывающие социальную активность в период ивента (обсуждения игры/ивента в соц.сетях, на формах).  Они помогут определить степень интереса к тематике ивента.  
   • Повышение/понижение рейтинга приложения.  
   • Сетевой эффект. Количество реферальных пользователей (если такое есть в продукте, конечно).  

### Метрики дохода

   • ARPU за период ивента — основной показатель монетизационной эффективности. Изменения покажут, стали ли доходы в общем больше или меньше.  
   • ARPPU за период ивента — реакция лояльных платящих пользователей.  
   • ARPDAU —  посмотреть, как менялись траты пользователей в разные дни ивента. Возможно, в начале были просадки, пока пользователи считали, что у них ещё достаточно времени (мб стоит делать какие-то акции для стимуляции продаж в первые дни?)  
   • Процент платящих пользователей — стимулировало ли событие активных, но не платящих пользователей потратить деньги ради контента ивента.  
   • Средний чек. Разница в среднем чеке и ARPPU покажет, много ли повторных (относительно первого в период ивента) платежей делают пользователи.  

Все эти метрики обязательно смотреть в разрезе сегментов пользователей (гео, уровень, пол и т.д.), чтобы понять на кого проведение ЭТОГО ивента повлияло в большей или меньшей степени.  
При анализе нужно учитывать, были ли какие-то другие изменения в продукте или акции, которые могли повлиять на ARPU наряду с ивентом.  


## Вопросы на подумать

   • Был ли спад/рост доходов после ивента? Возможно, в сам ивент мы зарабатывали больше, чем в обычный период, но потом нет ли спада т.к. игроки "выдохлись"? Не нивелировало ли это весь полезный эффект?  
   • Понять "кривую" сложности уровней. В идеале сложность должна повышаться и понижаться. Напоминать бег с барьерами. Чтобы давать игрокам и чувство "челленжа", и вдохновлять успехами. Возможно, на каких-то уровнях использовались новые механики, стоит посмотреть легко или сложно они дались пользователям.  
   • Изменилось ли время, проведённое в игре в общем? Просело ли время игроков в основной игре?  
   • Изменение метрик дохода у разных сегментов пользователей. Позволит понять, среди каких сегментов и как проявил себя ивент.   Был ли успешен у "китов" или повысил доходы с "пескарей"?

## Часть 2: изменение метрик при усложнении механики ивента 

Предположим, в другом событии мы усложнили механику событий так, что при каждой неудачной попытке выполнения уровня игрок будет откатываться на несколько уровней назад. Изменится ли набор метрик оценки результата? Если да, то как?
________

В целом ядро ивента остаётся прежним. Основные метрики остаются в силе.
Метрики дохода пользователей остаются в силе, т.к. это важная часть всего проекта.  

Т.к. изменения в механике направлены на усложнение игры, логично предположить, что наша цель стимулировать пользователя к покупке бонусов для преодоления этого усложнения. Тут, как говорится, главное не переборщить и не отбить у пользователей интерес к прохождению после поражений. Поэтому тут будет особенно важно пристально следить не только за метриками дохода, но и за метриками вовлечённости в игру.
 
Дополнительно можно посмотреть:  
1. На каких уровнях игроки проигрывали больше всего.
2. Сравнить отток пользователей на разных уровнях с прошлыми ивентами без усложнения механики. Увеличился ли отток при усложнении? 
4. После скольки попыток взять уровень игрок прекращал играть? 
5. Тратил и покупал ли он больше бонусов после отката? На каком уровне? Сразу после отката или на том, откуда откатился?
